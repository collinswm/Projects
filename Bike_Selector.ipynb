## **Identify Your Racebike**
---

The objective is to train a Convolutional Neural Network to recognise the brand of racebike based on preloaded images covering 13 classes representing major manufacturers.

Becasue each of these bikes follows a very simimlar design aesthetic - mainly driven by aerodynamics and UCI rules on size and dimensions of frames there is very little to distinguish each bike.

Early in the model you can see there is a very high error rate. However the process followed was to progressively retrain the model with higher epochs and a more refined learning rate to see if we can reach a stable / good enough model.

As can be seen at the end, random images thrown at the resnet34 model are accurately identified.

#### Apply ImageDataBunch to Create Training and Validation Sets


---



Then apply ImageDataBunch method to 
(1) create a training data path 
(2) split the data into training 80% and validation 20%, 
(3) transform the images into size 224 which appears to be a good default and 
(4)then run a normalise procedure.

```
# The ImageDataBunch is a critical function in the Pytorch/Fastai library - come back to it later. 

np.random.seed(42)
data = ImageDataBunch.from_folder(path, 
                                  train=".", 
                                  valid_pct=0.2, # Identify images, create a training and validation set with 80% - 20% split
                                  ds_tfms=get_transforms(), # Later on I will specify do_flip, flip_vert, max_rotate, max_warp etc
                                  size=224, # Here you determine the size of the images. Lower resolution possible if GPU capacity overload
                                  num_workers=4).normalize() # Normalize `x` with `mean` and `std` I am choosing not normalise here 
```

## Train the Model


---



Step 1: Build the CNN learner using Resnet and setting the error rate as the key metric. This can be changed to Resnet50 or another network architecutre later on

Step 2: Run the model using "Fit One Cycle" and doing it against 4 epochs. Again this can be altered later on with more or fewer epochs.

Step 3: Save the model as a checkpoint as "topwood_stage1". This makes it more efficient later when reloading. 

Step 4: Unfreeze the weights and bias in the model. This will be unpacked as the models become more sophisticated - for now I am doing it all off the shelf

Step 5: Now I run the lr_find() function to try to deterine the optimal learning rate so that the stochasitc gradient decent are not too high (overshoot) and not too low (get stuck with a local minimum) 

Step 6: We will fit the model a few more times and mess around with the learning rate and epoch sizes to see if we can improve error rates. Generally I think there will be some refinement of images needed because the error rates are high. See below.



**Create Learner **

```
# Build the learner using a Convolutional Neural Network

learn = cnn_learner(data, models.resnet34, metrics=error_rate) # Later if I can get this thing sorted I will go back and do a ResNet50 model
```

**Fit Model**

```
# Here I am running the first learner. 
# It will use "Fit One Cycle" which is current best practice.
learn.fit_one_cycle(4)
```

**Save Model**

```
# I am saving this first learn iteration as topwood_stage1. 
# It can be reloaded from here without having to go through preliminary steps

learn.save('stage1-4ep')
```

**Scan Results**

```
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

Plot Top Losses

There is a fair amount of confusion in the model when it comes to the default settings. We will really need to get stuck into the training - more epochs and better learning rates

```
interp.plot_top_losses(9, figsize=(15,11))
```

Confusion Matrix
Clearly there are a lot of incorrect classifications. Further training required. Maybe in later scenario we need to take a look at resnet50

```
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

**Model is most confused about the following bike brands**

```
interp.most_confused(min_val=2)
```

#### Unfreeze, Fine Tune and Identify Learning Rates

---



```
learn.unfreeze()
```

```
learn.fit_one_cycle(1)
```

```
learn.load('stage1-4ep');
```

**Find Learning Rate**

```
learn.lr_find() # Here I am trying to find the point at which the learning rate starts to curve upwards. Then step back to high point of slope for ooptimal learning rate
```

**Visualise Learning Rate**

```
# If the plot is not showing try to give a start and end learning rate
# learn.lr_find(start_lr=1e-5, end_lr=1e-1)

learn.recorder.plot()

# I will check out LR 1e-5 - 1e-4
```

**Run New Cycle**

```
# 5 Epoch LR 6-4

learn.fit_one_cycle(5, max_lr=slice(1e-6, 1e-4))

# Using the new learning rates the errors are still very very high. 
# This setting does not help the error rates. I will change learning rate and epochs in further cycles
# LR could be causing a problem with local minimum - there is a dip after the 1e-6
```

**Save Cycle**

```
# Here I will run a new fit one cycle, with default learning rates. So no reference to the chart above. 
learn.save('stage2-5epoch_lr64')
```

**Run New Cycle**

```
# 6 Epoch - LR 5-4

learn.fit_one_cycle(6, max_lr=slice(1e-5, 1e-4))

## Some improvemen in learning rates. I will iterate below
```

**Save Cycle**

```
learn.save('stage3-6epoch-lr54')
```

**Run New Cycle**

```
# 8 epoch and LR 5-3

learn.fit_one_cycle(8, max_lr=slice(1e-5, 1e-3))

# Once again a significant improvement in the error rates and training losses. 
# I will continue to train the model until the training loss starts to move upward
# I suspect the Learning Rate is a big driver for improvement - better slope
```

**Save Cycle**

```
learn.save('stage4-8epoch-lr5-3')
```

**Run New Cycle**

```
# 10 Epoch LR 5-3

learn.fit_one_cycle(10, max_lr=slice(1e-5, 1e-3))

#Here we see that the training losses are still dropping and error rates are not at 40. But are we overtraining?
```

**Save Cycle**

```
learn.save('stage5-10epoch-lr53')
```

Run New Cycle

```
# 12 Epoch LR 5-3

learn.fit_one_cycle(12, max_lr=slice(1e-5, 1e-3))
```

**Save Cycle**

```
learn.save('stage6-12epoch-lr53')
```

**Run New Cycle**

```
# 15 Epoch LR 6-3 (broaden the learning rate - to check)

learn.fit_one_cycle(15, max_lr=slice(1e-6, 1e-3))
```

**Save Cycle**

```
learn.save('stage7-15epoch-lr63')
```

Resnet50 usually performs better because it is a deeper network with more parameters. Let's see if we can achieve a higher performance here. To help it along, let's us use larger images too, since that way the network can see more detail. We reduce the batch size a bit since otherwise this larger network will require more GPU memory.
