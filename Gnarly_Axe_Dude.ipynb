```
# Create Classes

classes = ['es335', 'ibanezjem', 'jaguar', 'lespaul', 'prs', 'stratocaster', 'telecaster', 'rickenbacker360', 'dangelico', 'sg', 'gretsche6128']
print(array(classes))
len(classes)
```

```
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path, 
                                  train=".", 
                                  valid_pct=0.2, # Identify images, create a training and validation set with 80% - 20% split
                                  ds_tfms=get_transforms(), # Later on I will specify do_flip, flip_vert, max_rotate, max_warp etc
                                  size=224, # Here you determine the size of the images. Lower resolution possible if GPU capacity overload
                                  num_workers=4).normalize() # Normalize `x` with `mean` and `std` I am choosing not normalise here 
```

```
data.classes
```

```
# Check classes, dimensions and size of Training and Validation

data.classes, data.c, len(data.train_ds), len(data.valid_ds) 
```

```
data.show_batch(rows=10, figsize=(20, 15))
```

```
# Build the learner using a Convolutional Neural Network

learn = cnn_learner(data, models.resnet34, metrics=error_rate) # Later if I can get this thing sorted I will go back and do a ResNet50 model
```

```
# It will use "Fit One Cycle" which is current best practice.
learn.fit_one_cycle(4)
```

```
learn.save('stage1')
```

```
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```
interp.most_confused(min_val=2)
```
